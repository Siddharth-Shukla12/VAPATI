```python
import pandas as pd
```

```python
crop = "../Data-raw/cpdata.csv"
fert = "../Data-raw/fertilizer.csv"
```

```python
crops = pd.read_csv(crop)
fertilizer = pd.read_csv(fert)
```

```python
crops.head()
```

```python
fertilizer.head()
```

```python
def change(x):
    x = x.replace(" ", "")
    x = x.lower()
    return x
```

```python
crops['label'] = crops['label'].apply(change)
fertilizer['Crop'] =  fertilizer['Crop'].apply(change)
```

```python
fertilizer['Crop'] = fertilizer['Crop'].replace('mungbeans','mungbean')
fertilizer['Crop'] = fertilizer['Crop'].replace('lentils(masoordal)','lentil')
fertilizer['Crop'] = fertilizer['Crop'].replace('pigeonpeas(toordal)','pigeonpeas')
fertilizer['Crop'] = fertilizer['Crop'].replace('mothbean(matki)','mothbeans')
fertilizer['Crop'] = fertilizer['Crop'].replace('chickpeas(channa)','chickpea')
```
